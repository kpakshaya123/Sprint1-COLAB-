*Remove* Rows Which has Missing Values


```python
df=df.dropna()
```

Replace The Number By String("BMW")

```python
df['Model']=df['Model'].apply (lambda x:x if str(x).isalpha() else "BMW")
```

```python
df.head(200)
```

Handling Invalid Values in Year Column

1)Convert the Column to Numeric Value

```python
df['Year']=pd.to_numeric(df['Year'] ,errors="coerce")
```

2)Invalid Year values are replaced with NaN.

```python
df.loc[df['Year'] < 1000, 'Year'] = pd.NA
```

3)Rows where the Year value is missing (NaN) are removed.

```python
df=df.dropna(subset=['Year'])
```

4)It keeps only rows where the Year is a whole number and removes decimal values.

```python
df=df[df['Year']%1==0]
```
